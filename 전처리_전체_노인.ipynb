```python
import pandas as pd
```

```python
df = pd.read_csv('성별_전체_추계인구.csv', encoding='euc-kr')
```

```python
# 80세이상 - 를 0으로 변환
df.replace('-', 0, inplace=True)
```

```python
# 모든 숫자 열을 정수형으로 변환
for col in df.columns[2:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# '계' 데이터를 필터링하여 별도로 저장
total_population = df[df['연령별'] == '계']
```

```python
total_population
```

```python
# 65세 미만 인구 데이터 필터링 (0-64세)
under_65_data = df[df['연령별'].str.contains('^(0 - 4세|5 - 9세|10 - 14세|15 - 19세|20 - 24세|25 - 29세|30 - 34세|35 - 39세|40 - 44세|45 - 49세|50 - 54세|55 - 59세|60 - 64세)$')]
```

```python
under_65_data
```

```python
# 65세 미만 인구 합산
under_65_population = under_65_data.drop(columns=['성별']).set_index('연령별').sum().to_frame().T
under_65_population['성별'] = '전체'
under_65_population['연령별'] = '노년x'
```

```python
# 65세 이상 인구 데이터 필터링 (65세 이상)
over_65_data = df[df['연령별'].str.contains('^(65 - 69세|70 - 74세|75 - 79세|80 - 84세|85 - 89세|90 - 94세|95 - 99세|100세 이상)$')]
```

```python
over_65_data
```

```python
# 65세 이상 인구 합산
over_65_population = over_65_data.drop(columns=['성별']).set_index('연령별').sum().to_frame().T
over_65_population['성별'] = '전체'
over_65_population['연령별'] = '노년'
```

```python
over_65_population
```

```python
# '계' 데이터와 합산된 데이터를 결합
combined_population = pd.concat([total_population, under_65_population, over_65_population])


# 결과 출력
print(combined_population)

# 결과 저장
combined_population.to_csv('성별_전체_추계인구_전처리1.csv', index=False, encoding='euc-kr')
```
