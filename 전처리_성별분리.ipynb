```python
import pandas as pd
```

```python
df = pd.read_csv('성_및_연령별_추계인구_1세별__5세별____전국_20240531154232.csv', encoding='euc-kr')
```

```python
df = df.drop(columns=['가정별'])
```

```python
print(df.columns)
```

```python
df.head
```

```python
# 성별로 데이터 분리
df_total = df[df['성별'] == '전체']
df_male = df[df['성별'] == '남자']
df_female = df[df['성별'] == '여자']
```

```python
# CSV 파일로 저장
df_total.to_csv('성별_전체_추계인구.csv', index=False, encoding='euc-kr')
df_male.to_csv('성별_남자_추계인구.csv', index=False, encoding='euc-kr')
df_female.to_csv('성별_여자_추계인구.csv', index=False, encoding='euc-kr')
```
